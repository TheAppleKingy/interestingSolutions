```python
#Формат входных данных
#На вход программе подаются две строки текста, содержащие по одному слову из перечня "камень", "ножницы", "бумага", "ящерица" или "Спок". 
#На первой строке записан выбор Тимура, на второй – выбор Руслана.

#Формат выходных данных
#Программа должна вывести результат жеребьёвки: кто победил - Тимур или Руслан, или они сыграли вничью.


timur=input()
ruslan=input()
if timur==ruslan:
    print('ничья')
else:
    камень={'ножницы':True,
            'бумага':False,
            'ящерица':True,
            'Спок':False}
    ножницы={'камень':False,
            'бумага':True,
            'ящерица':True,
            'Спок':False}
    бумага={'камень':True,
            'ножницы':False,
            'ящерица':False,
            'Спок':True}
    спок={'камень':True,
            'ножницы':True,
            'бумага':False,
            'ящерица':False}
    ящерица={'Камень':False,
            'ножницы':False,
            'Спок':True,
            'бумага':True}
    g=''
    T={'камень':камень,
        'ножницы':ножницы,
        'бумага':бумага,
        'Спок':спок,
        'ящерица':ящерица}
    for rus in T[timur]:
        if rus==ruslan and T[timur][rus]==True:
            g='Тимур'
    if g=='':
        g='Руслан'
    print(g)
```
