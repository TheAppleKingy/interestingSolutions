```python
#Необходимо написать программу, реализующую алгоритм написания этой песни. Алгоритм выводит в конце предложения следующую в алфавитном порядке букву,
# если она встречается в строке текста, а очередную строку отображает уже без этой буквы.

#Формат входных данных
#На вход программе подается одно слово, записанное строчными русскими буквами без буквы "ё"

#Формат выходных данных
#Программа должна вывести в соответствии с указанным алгоритмом строки,
#количество которых равно количеству разных букв в строке, которая получается путем конкатенации введенного слова и строки "запретил букву".

word=input()
k=' запретил букву '
c=word+k
q=sorted(set(c))
q.remove(' ')
c=c+q[0]
for i in range(len(q)-1):
    if q[i] in c:
        print(c)
        word=word.translate({ord(q[i]):None})
        k=k.translate({ord(q[i]):None})
        c=word+k+q[i+1]   
        c=c.lstrip().replace('  ',' ')
c=c.replace('  ',' ')
print(c)
        

```
